```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
%matplotlib inline
```

```python
def installs_to_num(s):
    if s.endswith("M"):
        s = int(s.split(".")[0]) * 1000
    else:
        s = int(s.split(".")[0])
    return s


data = pd.read_csv("./android-games.csv")
data["category"] = data["category"].map(lambda x: x[5:])
data["installs"] = data["installs"].map(installs_to_num)
data.head()
```

```python
g = data.groupby("category").mean()
g.head()
```

```python
alt.Chart(g.reset_index()).mark_bar().encode(
    x="total ratings:Q",
    y=alt.Y("category:N", sort="-x"),
    color=alt.Color("category:N", legend=None)
)
```
